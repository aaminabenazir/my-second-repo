```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


tips = sns.load_dataset("tips")
print(tips.head())
```

Task 1 — Your First Matplotlib Chart

```python
daily_tips = tips.groupby('day', observed=False)['tip'].sum().reset_index()

plt.figure(figsize=(8,5))
plt.bar(daily_tips['day'], daily_tips['tip'], color="green",
         linewidth=2,edgecolor="black")
plt.title("Tips by Day")
plt.xlabel("Days of the Week")
plt.ylabel("Sum of Tips")
plt.tight_layout()
plt.show()
```

Task 2 — Seaborn Histogram

```python
sns.histplot(tips, x="tip", bins=20, kde=True, color="red",edgecolor="black")
plt.title("Distribution of Tip Amounts")
plt.xlabel("Tip Amount")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

Task 3 — Seaborn Scatter Plot

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day", style="time",edgecolor="black")
plt.title("Total Bill vs Tip Amount")
plt.xlabel("Total Bill")
plt.ylabel("Tip ")
plt.tight_layout()
plt.show()
```
