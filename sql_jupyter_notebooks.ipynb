Task 2 — WHERE vs. HAVING

```python
sql_query_string ="""SELECT CustomerID, count(Freight) AS high_freight_orders
FROM orders
where freight>50
GROUP BY CustomerID;

"""
countof_freight = pd.read_sql(sql_query_string, conn)

print(countof_freight)
```

```python
sql_query_string ="""SELECT CustomerID, SUM(Freight) AS total_freight
FROM orders
GROUP BY CustomerID
having SUM(Freight)>500;

"""
countof_freight = pd.read_sql(sql_query_string, conn)

print(countof_freight)
```

Task 3 — JOIN and Aggregation

```python
sql_query_string ="""SELECT customers.CompanyName, customers.Country, COUNT(orders.OrderID) AS total_no_of_orders, SUM(orders.Freight) AS total_freight
FROM customers
INNER JOIN orders ON customers.CustomerID = orders.CustomerID
GROUP BY customers.CustomerID, customers.CompanyName, customers.Country;
"""
inner_join= pd.read_sql(sql_query_string, conn)

print(inner_join)
```

```python
sql_query_string ="""SELECT customers.CompanyName, customers.Country, COUNT(orders.OrderID) AS total_no_of_orders, SUM(orders.Freight) AS total_freight
FROM customers
left JOIN orders ON customers.CustomerID = orders.CustomerID
GROUP BY customers.CustomerID, customers.CompanyName, customers.Country;
"""
left_join= pd.read_sql(sql_query_string, conn)

print(left_join)
```

question:
Display both results and in a markdown cell, explain in 2–3 sentences what changed between the two queries and why.



ans :

INNER JOIN
1.Returns only the rows where there is a match in both tables based on the join condition.
2.Excludes non-matching rows from both tables.

 LEFT JOIN
1.Returns all rows from the left table, and the matching rows from the right table.

2.If there’s no match, columns from the right table will be NULL.
3. Keeps all left table data, even without matches.
